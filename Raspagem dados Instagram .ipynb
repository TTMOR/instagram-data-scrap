## Estraindo o perfil

## Filtragem 

## Filtrar somente por imagens

````python
import tkinter as tk
from tkinter import filedialog
import json

# Fun√ß√£o para escolher o arquivo JSON
def escolher_arquivo():
    arquivo = filedialog.askopenfilename(filetypes=[("Arquivos JSON", "*.json")])
    if arquivo:
        return arquivo
    return None

# Fun√ß√£o para carregar o JSON
def carregar_json(arquivo_json):
    with open(arquivo_json, 'r', encoding='utf-8') as f:
        return json.load(f)

# Fun√ß√£o para filtrar por imagens
def filtrar_imagens(dados_json):
    imagens = []
    
    # Iterar pelas postagens
    for postagem in dados_json['data']['xdt_api__v1__feed__user_timeline_graphql_connection']['edges']:
        node = postagem['node']
        
        # Verificar se a postagem tem imagens
        if 'image_versions2' in node and 'candidates' in node['image_versions2']:
            imagens.append(node)
    
    return imagens

# Fun√ß√£o principal para processar os dados
def processar_json():
    # Passo 1: Escolher o arquivo JSON
    arquivo_json = escolher_arquivo()
    if not arquivo_json:
        print("Nenhum arquivo selecionado.")
        return
    
    # Passo 2: Carregar o JSON
    dados_json = carregar_json(arquivo_json)
    
    # Passo 3: Filtrar por imagens
    imagens = filtrar_imagens(dados_json)
    print(f"Imagens encontradas: {len(imagens)}")
    
    # Exibir as imagens ou realizar outras a√ß√µes
    for imagem in imagens:
        print(imagem)

# Rodar o processo
processar_json()
````

## Extrair as curtidas e mostrar URL da Foto( Todos que crtiram foi removido do instagram) 

````python
# Fun√ß√£o para extrair curtidas e URLs
def extrair_curtidas_e_detalhes(imagens):
    imagens_com_detalhes = []
    
    for imagem in imagens:
        # N√∫mero de curtidas
        curtidas = imagem.get('like_count', 0)
        
        # Pessoas que curtiram (nomes)
        pessoas_que_curtiram = [
            usuario['node']['username']
            for usuario in imagem.get('liked_by', {}).get('edges', [])
        ]
        
        # URL da imagem
        url_imagem = imagem.get('image_versions2', {}).get('candidates', [{}])[0].get('url', 'Sem URL')
        
        # Adicionar informa√ß√µes detalhadas
        imagens_com_detalhes.append({
            'caption': imagem.get('caption', {}).get('text', 'Sem legenda'),
            'curtidas': curtidas,
            'url_imagem': url_imagem,
            'pessoas_que_curtiram': pessoas_que_curtiram
        })
    
    return imagens_com_detalhes

# Fun√ß√£o principal atualizada para exibir URLs e pessoas que curtiram
def processar_json():
    # Passo 1: Escolher o arquivo JSON
    arquivo_json = escolher_arquivo()
    if not arquivo_json:
        print("Nenhum arquivo selecionado.")
        return
    
    # Passo 2: Carregar o JSON
    dados_json = carregar_json(arquivo_json)
    
    # Passo 3: Filtrar por imagens
    imagens = filtrar_imagens(dados_json)
    print(f"Imagens encontradas: {len(imagens)}")
    
    # Passo 4: Extrair detalhes
    imagens_com_detalhes = extrair_curtidas_e_detalhes(imagens)
    
    # Passo 5: Encontrar a imagem com mais curtidas
    imagem_mais_curtida = max(imagens_com_detalhes, key=lambda x: x['curtidas'])
    
    # Exibir os detalhes da imagem com mais curtidas
    print(f"Imagem com mais curtidas: {imagem_mais_curtida['caption']}")
    print(f"N√∫mero de curtidas: {imagem_mais_curtida['curtidas']}")
    print(f"URL da imagem: {imagem_mais_curtida['url_imagem']}")
    print(f"Pessoas que curtiram: {imagem_mais_curtida['pessoas_que_curtiram']}")

# Rodar o processo
processar_json()
````

````python
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt

# Fun√ß√£o para exibir a imagem diretamente no Jupyter Notebook
def exibir_imagem_no_jupyter(url):
    try:
        # Fazendo o download da imagem
        response = requests.get(url)
        response.raise_for_status()
        imagem = Image.open(BytesIO(response.content))
        
        # Exibindo a imagem
        plt.figure(figsize=(8, 8))
        plt.imshow(imagem)
        plt.axis('off')  # Remove os eixos
        plt.show()
    except Exception as e:
        print(f"Erro ao exibir a imagem: {e}")

# Atualizar o fluxo principal para plotar no Jupyter
def processar_json_plot():
    # Passo 1: Escolher o arquivo JSON
    arquivo_json = escolher_arquivo()
    if not arquivo_json:
        print("Nenhum arquivo selecionado.")
        return
    
    # Passo 2: Carregar o JSON
    dados_json = carregar_json(arquivo_json)
    
    # Passo 3: Filtrar por imagens
    imagens = filtrar_imagens(dados_json)
    print(f"Imagens encontradas: {len(imagens)}")
    
    # Passo 4: Extrair detalhes
    imagens_com_detalhes = extrair_curtidas_e_detalhes(imagens)
    
    # Passo 5: Encontrar a imagem com mais curtidas
    imagem_mais_curtida = max(imagens_com_detalhes, key=lambda x: x['curtidas'])
    
    # Exibir os detalhes da imagem com mais curtidas
    print(f"Imagem com mais curtidas: {imagem_mais_curtida['caption']}")
    print(f"N√∫mero de curtidas: {imagem_mais_curtida['curtidas']}")
    print(f"URL da imagem: {imagem_mais_curtida['url_imagem']}")
    print(f"Pessoas que curtiram: {imagem_mais_curtida['pessoas_que_curtiram']}")
    
    # Exibir a imagem no Jupyter
    print("Exibindo a imagem com mais curtidas no Jupyter...")
    exibir_imagem_no_jupyter(imagem_mais_curtida['url_imagem'])

# Rodar o processo
processar_json_plot()
````

## iterando entre cada postagems e fotos 

````python
import json
import IPython.display as display
from PIL import Image
import requests
from io import BytesIO
from tkinter import Tk, filedialog


# Fun√ß√£o para carregar o JSON
def carregar_json(caminho_arquivo):
    with open(caminho_arquivo, 'r', encoding='utf-8') as arquivo:
        dados = json.load(arquivo)
    return dados


# Fun√ß√£o para exibir uma postagem (imagem, curtidas e coment√°rios)
def exibir_postagem(postagem):
    print("Legenda:", postagem.get('caption', {}).get('text', 'Sem legenda'))
    print("N√∫mero de curtidas:", postagem.get('like_count', 0))

    # Tratamento para evitar erro quando n√£o h√° coment√°rios
    comentarios = postagem.get('comments') or {}
    comentarios_data = comentarios.get('data', [])
    print("Coment√°rios:")
    if comentarios_data:
        for comentario in comentarios_data:
            print("-", comentario.get('text', ''))
    else:
        print("Sem coment√°rios.")

    # Exibir a imagem da postagem
    url_imagem = postagem.get('image_versions2', {}).get('candidates', [{}])[0].get('url')
    if url_imagem:
        response = requests.get(url_imagem)
        img = Image.open(BytesIO(response.content))
        display.display(img)
    else:
        print("Imagem n√£o encontrada.")


# Fun√ß√£o principal para iterar pelas postagens
def navegar_postagens(postagens):
    index = 0

    while True:
        print(f"\nPostagem {index + 1} de {len(postagens)}:")
        exibir_postagem(postagens[index])

        acao = input("Digite [n] para pr√≥xima, [p] para anterior, ou [s] para sair: ").strip().lower()
        if acao == 'n':
            if index < len(postagens) - 1:
                index += 1
            else:
                print("Voc√™ j√° est√° na √∫ltima postagem.")
        elif acao == 'p':
            if index > 0:
                index -= 1
            else:
                print("Voc√™ j√° est√° na primeira postagem.")
        elif acao == 's':
            print("Encerrando navega√ß√£o.")
            break
        else:
            print("Comando inv√°lido. Tente novamente.")


# Fun√ß√£o para abrir o seletor de arquivos com Tkinter
def selecionar_arquivo():
    root = Tk()
    root.withdraw()
    caminho_arquivo = filedialog.askopenfilename(title="Selecione o arquivo JSON", filetypes=[("Arquivos JSON", "*.json")])
    return caminho_arquivo


# In√≠cio do programa
caminho_arquivo = selecionar_arquivo()
if caminho_arquivo:
    dados = carregar_json(caminho_arquivo)
    postagens = dados.get('data', {}).get('xdt_api__v1__feed__user_timeline_graphql_connection', {}).get('edges', [])
    postagens = [p['node'] for p in postagens]  # Extrair os n√≥s das postagens
    navegar_postagens(postagens)
else:
    print("Nenhum arquivo selecionado.")
````

## Ou podemos usar o hist√≥rico, ou carregar a lista de v√°rios. 

````python
import instaloader
import os
import json
import time
import tkinter as tk
from tkinter import filedialog

# Fun√ß√£o para escolher o diret√≥rio
def escolher_diretorio():
    root = tk.Tk()
    root.withdraw()  # Esconde a janela principal
    pasta = filedialog.askdirectory(title="Escolha o diret√≥rio para salvar os arquivos")
    return pasta

# Fun√ß√£o para iniciar o scraping e salvar os dados
def realizar_scraping(diretorio, instagram_handle):
    # Criando uma inst√¢ncia do Instaloader
    L = instaloader.Instaloader()

    # Custom User-Agent para evitar bloqueios
    L.context._user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    
    # Nome do perfil que voc√™ quer buscar
    instagram_handle = 'nome_de_usuario'  # Substituir pelo perfil desejado

    # Carregar o perfil do Instagram
    profile = instaloader.Profile.from_username(L.context, instagram_handle)

    # Preparar uma lista para armazenar os dados raspados
    data = []

    # Iterar sobre cada post no perfil
    for post in profile.get_posts():
        post_data = {
            'PostID': post.shortcode,
            'Likes': post.likes,
            'Comments': post.comments,
            'Shares': 'N/A'  # Instagram API n√£o fornece contagem de compartilhamentos diretamente
        }
        data.append(post_data)
        
        # Adicionar um atraso de 20 segundos entre as solicita√ß√µes para evitar limita√ß√£o de taxa
        time.sleep(20)

    # Salvar os dados em um arquivo JSON
    json_file = os.path.join(diretorio, f"{instagram_handle}_instagram_data.json")
    with open(json_file, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)
    
    print(f"Data for {instagram_handle} has been saved to {json_file}")

# Escolher o diret√≥rio antes de come√ßar o scraping
diretorio = escolher_diretorio()

# Certificar-se de que um diret√≥rio foi escolhido
if diretorio:
    # Perguntar pelo nome do perfil do Instagram para raspar
    instagram_handle = input("Digite o nome do perfil do Instagram que voc√™ deseja raspar: ")

    # Realizar o scraping e salvar os dados
    realizar_scraping(diretorio, instagram_handle)
else:
    print("Nenhum diret√≥rio foi escolhido. O processo foi cancelado.")
````
